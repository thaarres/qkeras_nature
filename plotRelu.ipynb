```python
import numpy as np
from qkeras import quantized_relu
import matplotlib.pyplot as plt
import mplhep as hep
plt.style.use(hep.style.CMS)

fig = plt.figure(figsize=(15,10))
cols = ['#d55e00','#cc79a7','#0072b2','#f0e442','#009e73']
x = np.linspace(-1, 2, 100)
y1 = (x >= 0) * x
y2 = quantized_relu(3, 0)(x)
y3 = quantized_relu(3, 1)(x)
y4 = quantized_relu(2, 0)(x)
y5 = quantized_relu(6, 0)(x)
plt.plot(x, y1, label='ReLU',c=cols[0],linewidth=3)
plt.plot(x, y4, label='quantized_relu( bits=2, integer=0 )',c=cols[1],linewidth=3)
plt.plot(x, y2, label='quantized_relu( bits=3, integer=0 )',c=cols[4],linewidth=3)
plt.plot(x, y3, label='quantized_relu( bits=3, integer=1 )',c=cols[2],linewidth=3)
plt.plot(x, y5, label='quantized_relu( bits=6, integer=0 )',c=cols[3],linewidth=3)
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='best',frameon=False)
plt.show()
fig.tight_layout()
fig.savefig('relu.pdf')
```
